```python
url = 'https://data.boston.gov/api/3/action/datastore_search_sql'
```

```python
sql = '''
SELECT "DISTRICT",
       "OCCURRED_ON_DATE",
       "SHOOTING",
       "OFFENSE_DESCRIPTION"
FROM "b973d8cb-eeb2-4e7e-99da-c92938efc9c0"
WHERE CAST("SHOOTING" AS INT)=1
'''
```

```python
import requests, urllib
```

```python
final_url = url+"?sql="+urllib.parse.quote(sql)
```

```python
final_url
```

```python
rows = requests.get( final_url ).json()
```

```python
shootings = rows['result']['records']
```

```python
shootings[0]
```

```python
shootings[5:10]
```

```python
for s in shootings:
  print(s["DISTRICT"])
```

```python
import pandas as pd

df = pd.DataFrame(shootings)

df["OCCURRED_ON_DATE"] = pd.to_datetime(df["OCCURRED_ON_DATE"])
df["HOUR"] = df["OCCURRED_ON_DATE"].dt.hour
```

```python
hour_counts = df.groupby("HOUR").size().reset_index(name="count")
```

```python
import plotly.express as px

fig1 = px.line(hour_counts, x="HOUR", y="count",
               title="Shootings by Hour of Day")
fig1.show()
```

```python
peak_df = df[(df["HOUR"] >= 20) & (df["HOUR"] <= 23)]

crime_counts = peak_df.groupby("OFFENSE_DESCRIPTION").size().reset_index(name="count")
crime_counts = crime_counts.sort_values(by="count", ascending=False).head(5)
```

```python
fig2 = px.bar(crime_counts, x="OFFENSE_DESCRIPTION", y="count",
              title="Top Crimes During Peak Hours (8PM–11PM)")
fig2.show()
```
